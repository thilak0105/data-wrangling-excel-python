# XLSWritter

```python
import pandas as pd
import xlsxwriter

out_path = "/Users/thilak/PythonFiles/Sem 8/Data Wragling/Excel/xlsxwriter_output.xlsx"

# Write with formats and chart
with xlsxwriter.Workbook(out_path) as wb:
    ws = wb.add_worksheet("Report")
    header_fmt = wb.add_format({"bold": True, "bg_color": "#ADD8E6", "border": 1})
    num_fmt = wb.add_format({"num_format": "0.00", "border": 1})
    ws.write_row("A1", ["Name", "Score"], header_fmt)
    data = [("Alice", 85), ("Bob", 72), ("Charlie", 90), ("David", 65)]
    for r, (name, score) in enumerate(data, start=1):
        ws.write(r, 0, name)
        ws.write(r, 1, score, num_fmt)

    ws.conditional_format("B2:B5", {"type": "3_color_scale"})
    # Chart
    chart = wb.add_chart({"type": "column"})
    chart.add_series({"name": "Scores", "categories": "=Report!$A$2:$A$5", "values": "=Report!$B$2:$B$5"})
    ws.insert_chart("D2", chart)

df = pd.read_excel(out_path, sheet_name="Report")
print(df)
```
